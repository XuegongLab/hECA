```r
setwd('/stor/public/xixi/HCL')
meta <- read.csv('HCL_Fig1_cell_Info.csv', header = TRUE)
matrix <- read.table('Fetal-Skin3_rmbatchdge.txt', check.names = FALSE, header = TRUE, sep = ',')
matrix <- cbind(cell_ID = row.names(matrix), matrix)
dim(matrix)
matrix[1:10, 1:10]
```

```r
write.table(matrix, file = '/stor/public/hcad/skin_FetalSkin3_Guo2020/skin_FetalSkin3_Guo2020_expression.tsv', 
           sep = '\t', quote = FALSE, row.names = FALSE)
```

```r
cell_ID <- colnames(matrix)[-1]
meta_this <- meta[as.character(meta$batch)=='FetalSkin3', ]
dim(meta_this)
meta_this
```

```r
cell_cluster = rep("NA", length(cell_ID))
for (i in 1:length(cell_ID)){
    for (j in 1:dim(meta_this)[1]){
        if (cell_ID[i] == as.character(meta_this$cellnames[j])){
            cell_cluster[i] = as.character(meta_this$celltype[j])
            break
        }
    }
}
annotation = cbind(cell_ID = cell_ID, cell_cluster = cell_cluster)
annotation
```

```r
write.table(annotation, file = '/stor/public/hcad/skin_FetalSkin3_Guo2020/skin_FetalSkin3_Guo2020_annotation.tsv', 
           sep = '\t', quote = FALSE, row.names = FALSE)
```

```r
organ <- rep("skin", length(cell_ID))
region <- rep(as.character(meta_this$sample)[1], length(cell_ID))
subregion <- region
sample_status <- rep('normal', length(cell_ID))
donor_ID <- rep(as.character(meta_this$donor)[1], length(cell_ID))
donor_gender <- rep('female', length(cell_ID))
donor_age <- rep('11 weeks old', length(cell_ID))
metadata <- cbind(organ, region, subregion, sample_status, cell_ID, donor_ID, donor_gender, donor_age)
metadata
```

```r
write.table(metadata, file = '/stor/public/hcad/skin_FetalSkin3_Guo2020/skin_FetalSkin3_Guo2020_metadata.tsv', 
           sep = '\t', quote = FALSE, row.names = FALSE)
```
