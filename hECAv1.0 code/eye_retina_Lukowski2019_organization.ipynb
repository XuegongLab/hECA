```r
cell_ID <- colnames(dat)[-1]
cell_ID
```

```r
length(cell_ID)
```

```r
cell_cluster <- rep("NA", length(cell_ID))
anno <- cbind(cell_ID = cell_ID, cell_cluster = cell_cluster)
anno[1:10,]
```

```r
write.table(anno, file = "/stor/public/hcad/eye_retina_Lukowski2019/eye_retina_Lukowski2019_annotation.tsv", row.names = FALSE, quote = FALSE, sep = "\t")
```

```r
meta <- matrix(nrow = length(cell_ID), ncol = 8)
colnames(meta) <- c("organ", "region", "subregion", "sample_status", "cell_ID", "donor_ID", "donor_gender", "donor_age")
for (i in 1:nrow(meta)) {
    for (j in 1:ncol(meta)){
        if (j == 5) meta[i,j] = cell_ID[i]
        else if (j == 3) meta[i,j] = 'Neural retina'
        else if (j == 2) meta[i,j] = 'Retina'
        else if (j == 1) meta[i,j] = "Eye"
        else if (j == 4) meta[i,j] = "Normal"
        else meta[i,j] = "NA"
    }
}
meta[1:5,]
```

```r
dim(meta)
```

```r
write.table(meta, file = "/stor/public/hcad/eye_retina_Lukowski2019/eye_retina_Lukowski2019_metadata.tsv", row.names = FALSE, quote = FALSE, sep = "\t")
```
