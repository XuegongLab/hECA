```r
setwd('/stor/public/hcad/brain_ParietalTemporalFrontalFrontoparietal_Venteicher2017/')
library(dplyr)
mat <- read.delim("GSE89567_IDH_A_processed_data.txt", row.names = 1)
mat[1:5, 1:10]
```

```r
dim(mat)
```

```r
mattpm <- 2^mat[,]-1
```

```r
mattpm
```

```r
write.table(mattpm, file = "brain_ParietalTemporalFrontalFrontoparietal_Venteicher2017_expression.tsv", sep = "\t", quote = FALSE)
```

```r
anno <- rep('IDH-A', dim(mattpm)[2])
anno
```

```r
anno <- cbind(cell_ID = colnames(mat), cell_cluster = anno)
anno
```

```r
write.table(anno, file = "brain_ParietalTemporalFrontalFrontoparietal_Venteicher2017_annotation.tsv", row.names = FALSE, sep = "\t", quote = FALSE)
```

```r
meta <- matrix(nrow = nrow(anno), ncol = 8)
colnames(meta) <- c("organ", "region", "subregion", "sample_status", "cell_ID", "donor_ID", "donor_gender", "donor_age")
for (i in 1:nrow(meta)) {
    for (j in 1:ncol(meta)){
        if (j == 5) meta[i,j] = anno[i, 1]
        else if (j == 1) meta[i,j] = "brain"
        else if (j == 4) meta[i,j] = "glioma"
        else if (j == 6) {
            if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH42') meta[i,j] = 'Donor 1'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH43') meta[i,j] = 'Donor 2'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH44') meta[i,j] = 'Donor 3'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH45') meta[i,j] = 'Donor 4'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH56') meta[i,j] = 'Donor 5'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH57'||strsplit(anno[i,1], '_')[[1]][1] == 'X57') meta[i,j] = 'Donor 6'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH61') meta[i,j] = 'Donor 7'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH103'|| strsplit(anno[i,1], '_')[[1]][1] == 'mgh103') meta[i,j] = 'Donor 9'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH107pos'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107neg'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107') meta[i,j] = 'Donor 10'
            else meta[i,j] = 'Donor 8'
        }
        else if (j == 7) {
            if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH42') meta[i,j] = 'female'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH43') meta[i,j] = 'male'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH44') meta[i,j] = 'male'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH45') meta[i,j] = 'female'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH56') meta[i,j] = 'male'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH57'||strsplit(anno[i,1], '_')[[1]][1] == 'X57') meta[i,j] = 'male'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH61') meta[i,j] = 'male'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH103'|| strsplit(anno[i,1], '_')[[1]][1] == 'mgh103') meta[i,j] = 'female'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH107pos'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107neg'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107') meta[i,j] = 'female'
            else meta[i,j] = 'female'
        }
        else if (j == 8) {
            if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH42') meta[i,j] = '47'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH43') meta[i,j] = '45'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH44') meta[i,j] = '32'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH45') meta[i,j] = '32'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH56') meta[i,j] = '31'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH57'||strsplit(anno[i,1], '_')[[1]][1] == 'X57') meta[i,j] = '42'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH61') meta[i,j] = '24'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH103'|| strsplit(anno[i,1], '_')[[1]][1] == 'mgh103') meta[i,j] = '26'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH107pos'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107neg'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107') meta[i,j] = '47'
            else meta[i,j] = '40'
        }
        else if (j == 2) {
            if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH42')
                meta[i,j] = 'left parietal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH43')
                meta[i,j] = 'right temporal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH44')
                meta[i,j] = 'right frontal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH45')
                meta[i,j] = 'left frontal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH56')
                meta[i,j] = 'left parietal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH57'||strsplit(anno[i,1], '_')[[1]][1] == 'X57')
                meta[i,j] = 'right temporal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH61')
                meta[i,j] = 'right frontal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH103'|| strsplit(anno[i,1], '_')[[1]][1] == 'mgh103')
                meta[i,j] = 'left parietal'
            else if (strsplit(anno[i,1], '_')[[1]][1] == 'MGH107pos'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107neg'|| strsplit(anno[i,1], '_')[[1]][1] == 'MGH107')
                meta[i,j] = 'right frontal'
            else meta[i,j] = 'left frontoparietal'
        }
        else meta[i,j] = "NA"
    }
}
```
